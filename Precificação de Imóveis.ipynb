```python
dados [["Moeda","Preço","Freq"]] = dados["Valor"].str.split(expand = True)
dados.head()
```

```python
dados[dados["Freq"].isnull()]
```

```python
dados_filtrados = dados[dados["Freq"].isnull()]
dados_filtrados.head()
```

```python
dados_filtrados["Preço"].str.replace(".","").astype(float)
```

```python
dados_filtrados["Valor_anuncio_float"] = dados_filtrados["Preço"].str.replace(".","").astype(float)
```

```python
dados_filtrados.info()
```

```python
dados_filtrados["Valor_anuncio_float"].plot.hist(bins = 50)
```

```python
pd.set_option("display.precision", 2)
pd.set_option("display.float_format", lambda x: '%.2f' % x)
dados_filtrados.describe()
```

```python
sns.set()
plt.figure(figsize=(10,8))
sns.set_palette('Reds')
ax = sns.histplot(data = dados_filtrados, x = "Valor_anuncio_float", kde = True)
plt.xlim(0,10000000)
ax.set_title("Histograma Valor Imóveis")
plt.show()
```
